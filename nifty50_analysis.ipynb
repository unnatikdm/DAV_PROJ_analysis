# Import Required Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for plots
sns.set_style('whitegrid')
```

# Load the Clean Dataset

```python
df = pd.read_csv('data/cleaned_nifty50.csv')
df['Date'] = pd.to_datetime(df['Date'])
df.head()
```

# Inspect Dataset Shape and Structure

```python
print("Shape:", df.shape)
df.info()
```

# Check for Missing Values

```python
print("Missing values per column:")
print(df.isnull().sum())

# Visualize missing values
plt.figure(figsize=(10,6))
sns.heatmap(df.isnull(), cbar=False, yticklabels=False)
plt.title('Missing Values Heatmap')
plt.show()
```

# Summary Statistics

```python
print("Numerical Summary Statistics:")
display(df.describe())

print("\nCategorical Summary Statistics:")
display(df.describe(include='object'))
```

# Data Types and Unique Values

```python
print("Data Types:")
print(df.dtypes)

print("\nNumber of Unique Values per Column:")
print(df.nunique())

print("\nUnique Sectors:")
print(df['Sector'].unique())

print("\nSample Unique Tickers (first 10):")
print(df['Ticker'].unique()[:10])
```

# Correlation Analysis

```python
# Select numerical columns
num_cols = df.select_dtypes(include=[np.number]).columns
corr = df[num_cols].corr()

plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

# Data Distribution Visualization

```python
# Histograms
df[num_cols].hist(bins=50, figsize=(20,15))
plt.suptitle('Histograms of Numerical Features')
plt.show()

# Boxplots
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(20,10))
for i, col in enumerate(num_cols):
    ax = axes.flat[i]
    df[col].plot(kind='box', ax=ax)
    ax.set_title(col)
plt.suptitle('Boxplots of Numerical Features')
plt.tight_layout()
plt.show()
```

# Feature Relationships Visualization

```python
# Pairplot - sample to avoid large data
sample_df = df.sample(min(1000, len(df)))
sns.pairplot(sample_df[num_cols])
plt.suptitle('Pairplot of Numerical Features (Sample)')
plt.show()

# Scatter plot example: Open vs Close
plt.figure(figsize=(10,6))
plt.scatter(df['Open'], df['Close'], alpha=0.5)
plt.xlabel('Open')
plt.ylabel('Close')
plt.title('Scatter Plot: Open vs Close')
plt.show()
```

# Outlier Detection and Visualization

```python
# Outlier detection using IQR
def detect_outliers_iqr(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return (data < lower_bound) | (data > upper_bound)

# Count outliers for each numerical column
for col in num_cols:
    outliers = detect_outliers_iqr(df[col])
    print(f"Outliers in {col}: {outliers.sum()}")

# Visualize outliers in Close price
plt.figure(figsize=(10,6))
colors = ['blue' if not out else 'red' for out in detect_outliers_iqr(df['Close'])]
plt.scatter(df.index, df['Close'], c=colors, alpha=0.5)
plt.title('Outliers in Close Price (Red points are outliers)')
plt.show()
```
