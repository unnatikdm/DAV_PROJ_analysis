```python
# 1. Load Clean Data
# -------------------------------
df = pd.read_csv("cleaned_nifty50.csv")

df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values(by=['Ticker','Date'])
```

```python
# 2. Daily Returns
# -------------------------------
df['Daily_Return'] = df.groupby('Ticker')['Close'].pct_change()
```

```python
# 3. Moving Averages
# -------------------------------
df['MA_50'] = df.groupby('Ticker')['Close'].transform(lambda x: x.rolling(50).mean())
df['MA_100'] = df.groupby('Ticker')['Close'].transform(lambda x: x.rolling(100).mean())
```

```python
# 4. Volatility (20-day)
# -------------------------------
df['Volatility_20D'] = df.groupby('Ticker')['Daily_Return'].transform(lambda x: x.rolling(20).std())
```

```python
# 5. Trend Indicator (Bullish/Bearish)
# -------------------------------
df['Trend'] = df.apply(
    lambda row: 'Bullish' if row['MA_50'] > row['MA_100'] else 'Bearish',
    axis=1
)
```

```python
# 6. Cumulative Growth
# -------------------------------
df['Cumulative_Return'] = (1 + df['Daily_Return']).groupby(df['Ticker']).cumprod()
```
